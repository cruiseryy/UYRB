```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Myriad Pro'
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300

import numpy as np
from utils import *

# bin0 = np.loadtxt('binary.ref')
bin0 = np.loadtxt('../re1.reference')[:,-4:]

# tru0 = np.loadtxt('true.ref')
tru0 = np.loadtxt('../re0.reference')[:,-4:]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
mins_bin = np.min(bin0, axis=0)
maxs_bin = np.max(bin0, axis=0)
mins_tru = np.min(tru0, axis=0)
maxs_tru = np.max(tru0, axis=0)
mins = np.minimum(mins_bin, mins_tru)
maxs = np.maximum(maxs_bin, maxs_tru)
xlabels = ['obj1', 'obj2', 'obj3', 'obj4']
precision = [0, 1, 3, 2]
shading_tru = 0.0 * np.ones((tru0.shape[0], ))
shading_bin = 0.0 * np.ones((bin0.shape[0], ))

plotter = parallel_coords(mins, maxs, xlabels, precision)
plotter.plot_(ax, tru0, cmap = 'Reds_r', shading = shading_tru)
plotter.plot_(ax, bin0, cmap = 'Blues_r', shading = shading_bin)
```
