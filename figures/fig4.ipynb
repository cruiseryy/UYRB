```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Myriad Pro'
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
import matplotlib

import numpy as np

lyx_o0 = np.loadtxt('lyx_outflow0.csv', delimiter=',')
lyx_o1 = np.loadtxt('lyx_outflow1.csv', delimiter=',')
ljx_o0 = np.loadtxt('ljx_outflow0.csv', delimiter=',')
ljx_o1 = np.loadtxt('ljx_outflow1.csv', delimiter=',')
lyx_s0 = np.loadtxt('lyx_storage0.csv', delimiter=',') / 1e9 
lyx_s1 = np.loadtxt('lyx_storage1.csv', delimiter=',') / 1e9
ljx_s0 = np.loadtxt('ljx_storage0.csv', delimiter=',') / 1e9
ljx_s1 = np.loadtxt('ljx_storage1.csv', delimiter=',') / 1e9
```

```python
xx_o = np.arange(0, 324, 1)
```

```python
cmap_red = matplotlib.colormaps['Reds_r']
cmap_blue = matplotlib.colormaps['Blues_r']
cmap_grey = matplotlib.colormaps['Greys_r']
```

```python
np.sum(np.abs(lyx_s0[:,-1] - lyx_s1[:,-1]))
```

```python
xx_o = np.arange(0, 324, 1)
xx_s = np.arange(0, 325, 1)
fig, ax = plt.subplots(3, 1, figsize=(12, 7))
ax[0].plot(xx_o, lyx_o0[:, -2], label = 'Perfect', color = cmap_red(0.2), linewidth = 1.5)
ax[0].plot(xx_o, lyx_o1[:, -2], label = 'Binary', color = cmap_blue(0.2), linewidth = 1.5)
ax[0].plot(xx_o, lyx_o0[:, -1], label = 'History', color = cmap_grey(0.2), linewidth = 1.5)
for i in range(4):
    ax[0].plot(xx_o, lyx_o0[:, i], color = cmap_red(0.8), linewidth = 2, alpha = 0.8, zorder = 1)
    ax[0].plot(xx_o, lyx_o1[:, i], color = cmap_blue(0.8), linewidth = 2, alpha = 0.8, zorder = 1)
ax[0].legend()
ax[0].set_xlim([0, 324])
ax[0].set_xticks(np.arange(0, 325, 36))
ax[0].set_xticklabels(np.arange(2001, 2011, 1))

ax[1].plot(xx_o, ljx_o0[:, -2], label = 'Perfect', color = cmap_red(0.2), linewidth = 1.5)
ax[1].plot(xx_o, ljx_o1[:, -2], label = 'Binary', color = cmap_blue(0.2), linewidth = 1.5)
ax[1].plot(xx_o, ljx_o0[:, -1], label = 'History', color = cmap_grey(0.2), linewidth = 1.5)
for i in range(4):
    ax[1].plot(xx_o, ljx_o0[:, i], color = cmap_red(0.8), linewidth = 2, alpha = 0.8, zorder = 1)
    ax[1].plot(xx_o, ljx_o1[:, i], color = cmap_blue(0.8), linewidth = 2, alpha = 0.8, zorder = 1)
ax[1].set_xlim([0, 324])
ax[1].set_xticks(np.arange(0, 325, 36))
ax[1].set_xticklabels(np.arange(2001, 2011, 1))

ax[2].plot(xx_s, lyx_s0[:, -2] + ljx_s0[:, -2], label = 'Perfect', color = cmap_red(0.2), linewidth = 1.5)
ax[2].plot(xx_s, lyx_s1[:, -2] + ljx_s1[:, -2], label = 'Binary', color = cmap_blue(0.2), linewidth = 1.5)
ax[2].plot(xx_s, lyx_s0[:, -1] + ljx_s0[:, -1], label = 'History', color = cmap_grey(0.2), linewidth = 1.5)
for i in range(4):
    ax[2].plot(xx_s, lyx_s0[:, i] + ljx_s0[:, i], color = cmap_red(0.8), linewidth = 2, alpha = 0.8, zorder = 1)
    ax[2].plot(xx_s, lyx_s1[:, i] + ljx_s1[:, i], color = cmap_blue(0.8), linewidth = 2, alpha = 0.8, zorder = 1)
ax[2].set_xlim([0, 324])
ax[2].set_xticks(np.arange(0, 325, 36))
ax[2].set_xticklabels(np.arange(2001, 2011, 1))
```

```python
xx_o = np.arange(0, 324, 1)
xx_s = np.arange(0, 325, 1)

fig, ax = plt.subplots(3, 2, figsize=(10, 5), gridspec_kw={'height_ratios': [1, 1, 1], 'width_ratios': [0.05, 10]})
for i in range(3):
    ax[i][0].axis('off')
# Plot for the first subplot
ax[0][1].plot(xx_o, lyx_o0[:, -2], label='Perfect', color=cmap_red(0.2), linewidth=2)
ax[0][1].plot(xx_o, lyx_o1[:, -2], label='Binary', color=cmap_blue(0.2), linewidth=2)
ax[0][1].plot(xx_o, lyx_o0[:, -1], label='History', color=cmap_grey(0.2), linewidth=2)
for i in range(4):
    ax[0][1].plot(xx_o, lyx_o0[:, i], color=cmap_red(0.8), linewidth=2, zorder=1)
    ax[0][1].plot(xx_o, lyx_o1[:, i], color=cmap_blue(0.8), linewidth=2, zorder=1)
ax[0][1].set_xlim([0, 324])
ax[0][1].set_ylim([0, 1200])
ax[0][1].set_xticks([])  # Remove XTicks from the first subplot

# Plot for the second subplot
ax[1][1].plot(xx_o, ljx_o0[:, -2], label='Perfect', color=cmap_red(0.2), linewidth=2)
ax[1][1].plot(xx_o, ljx_o1[:, -2], label='Binary', color=cmap_blue(0.2), linewidth=2)
ax[1][1].plot(xx_o, ljx_o0[:, -1], label='History', color=cmap_grey(0.2), linewidth=2)
for i in range(4):
    ax[1][1].plot(xx_o, ljx_o0[:, i], color=cmap_red(0.8), linewidth=2, zorder=1)
    ax[1][1].plot(xx_o, ljx_o1[:, i], color=cmap_blue(0.8), linewidth=2, zorder=1)
ax[1][1].set_xlim([0, 324])
ax[1][1].set_ylim([0, 1800])
ax[1][1].set_xticks([])  # Remove XTicks from the second subplot

# Plot for the third subplot
ax[2][1].plot(xx_s, lyx_s0[:, -2] + ljx_s0[:, -2], label='Perfect', color=cmap_red(0.2), linewidth=2)
ax[2][1].plot(xx_s, lyx_s1[:, -2] + ljx_s1[:, -2], label='Binary', color=cmap_blue(0.2), linewidth=2)
ax[2][1].plot(xx_s, lyx_s0[:, -1] + ljx_s0[:, -1], label='History', color=cmap_grey(0.2), linewidth=2)
for i in range(4):
    ax[2][1].plot(xx_s, lyx_s0[:, i] + ljx_s0[:, i], color=cmap_red(0.8), linewidth=2, zorder=1)
    ax[2][1].plot(xx_s, lyx_s1[:, i] + ljx_s1[:, i], color=cmap_blue(0.8), linewidth=2, zorder=1)
ax[2][1].set_xlim([0, 324])
ax[2][1].set_xticks(np.arange(17, 325, 36))
ax[2][1].set_xticklabels(np.arange(2001, 2010, 1))
ax[2][1].xaxis.set_tick_params(length=0)

# Set shared Y-labels
fig.text(0.01, 0.68, 'Runoff [m$^3$/s]', va='center', rotation='vertical', fontsize=12)
fig.text(0.01, 0.22, 'Storage [bil m$^3$]', va='center', rotation='vertical', fontsize=12)

# Adjust layout
plt.tight_layout()

# Move the legend outside of the subplots
handles, labels = ax[0][1].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=3, bbox_to_anchor=(0.52, 1.06), fontsize=12)
plt.subplots_adjust(hspace=0.05)
plt.show()
```
