```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Myriad Pro'
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300

import numpy as np
from utils import *

bin0 = np.loadtxt('binary.ref')
bin1 = np.loadtxt('re1.reference')[:,-4:]

tru0 = np.loadtxt('true.ref')
tru1 = np.loadtxt('re0.reference')[:,-4:]
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
xx = tru0
yy = tru1

for i in range(4):
    ax[i//2, i%2].scatter(xx[:,i], yy[:,i], s=5)
    ax[i//2, i%2].axline([np.min(xx[:,i]), np.min(xx[:,i])],[np.max(xx[:,i]), np.max(xx[:,i])] , color='r')
    ax[i//2, i%2].set_xlabel('x')
    ax[i//2, i%2].set_ylabel('y')
    ax[i//2, i%2].set_title('obj {}'.format(i))
    print(np.max(np.abs(xx[:,i] - yy[:,i]))/np.min(np.abs(xx[:,i])))
```
